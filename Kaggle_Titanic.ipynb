```
import urllib.request 
import os
url="http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.xls"
filepath="titanic3.xls"
if not os.path.isfile(filepath):
    result=urllib.request.urlretrieve(url,filepath)
    print('downloaded:',result)
```

```
import numpy as np
import pandas as pd
data = pd.read_excel(filepath)
data.head(100)
```

```
cols = ['survived','pclass','sex','age','sibsp','parch','fare','embarked']  
data = data[cols]  
```

```
age_mean = data['age'].mean()
data['age'] = data['age'].fillna(age_mean)
fare_mean = data['fare'].mean()
data['fare'] = data['fare'].fillna(fare_mean)

data['sex']= data['sex'].map({'female':0, 'male': 1}).astype(int)

x_OneHot_data = pd.get_dummies(data=data,columns=["embarked" ])

ndarray = x_OneHot_data.values
Label = ndarray[:,0]
Features = ndarray[:,1:]

x_OneHot_data.head(100)
```

```
from sklearn import preprocessing
minmax_scale = preprocessing.MinMaxScaler(feature_range=(0, 1))
scaledFeatures=minmax_scale.fit_transform(Features)
```

```
msk = np.random.rand(len(scaledFeatures)) < 0.8  
x_train = scaledFeatures[msk]  
y_train = Label[msk]  
x_test = scaledFeatures[~msk]  
y_test = Label[~msk]  
```

```
from keras.models import Sequential  
from keras.layers import Dense
  
model = Sequential()  

model.add(Dense(units = 30, input_dim=9, kernel_initializer='uniform', activation='relu'))  

model.add(Dense(units = 60, kernel_initializer='uniform', activation='relu'))  

model.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid')) 

model.summary()  
```

```
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])  
train_history = model.fit(x_train, y_train, validation_split = 0.1, batch_size=30, epochs=20)  
```

```
score = model.evaluate(x_test, y_test)
print('loss:', score[0])
print('正確率', score[1])
```
